```python
import random
from datetime import datetime, timedelta

def gen_datetime(min_year=1900, max_year=datetime.now().year):
    # generate a datetime in format yyyy-mm-dd hh:mm:ss.000000
    start = datetime(min_year, 1, 1, 00, 00, 00)
    years = max_year - min_year + 1
    end = start + timedelta(days=365 * years)
    return start + (end - start) * random.random()
```

```python
def random_data():
    df_none = pd.DataFrame(columns=df_all.columns)
    for _ in range(len(df_all)*5):
        random_row = df_all.iloc[random.randint(0, 190071)].copy()
        random_row['DAY'] = gen_datetime(min_year=2012, max_year=2017).day
        random_row['MONTH'] = gen_datetime(min_year=2012, max_year=2017).month
        random_row['YEAR'] = gen_datetime(min_year=2012, max_year=2017).year
        random_row['DAY_WEEK'] = gen_datetime(min_year=2012, max_year=2017).weekday()
        random_row['HOUR'] = gen_datetime(min_year=2012, max_year=2017).hour
        random_row['LATITUDE'] = random_row['LATITUDE'] + random.uniform(-0.00001, 0.00001)
        random_row['LONGITUD'] = random_row['LONGITUD'] + random.uniform(-0.00001, 0.00001)
        df_none = df_none.append(random_row)
    return df_none
```

```python
df_none = random_data()
```
